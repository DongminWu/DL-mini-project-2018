```python
import pickle
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def save_np_array(np_list, filename):
    with open(filename, 'wb') as f:
        pickle.dump(list(np_list),f)

def restore_np_array(filename):
    with open(filename,'rb') as f:
        ret = pickle.load(f)
    return ret
```

```python
r_q_table = restore_np_array("./result_q_table.list")
len(r_q_table)
```

```python
r_q_network = restore_np_array("./result_q_learning.list")
len(r_q_network)
```

```python
r_policy = restore_np_array("./result_q_policy.list")
len(r_policy)
```

```python
def mean_reward(rList):
    ret_list = []
    for i in range(0,len(rList)-1, 100):
        ret = np.mean(rList[i:i+100])
        ret_list.append(ret)
        
    return np.array(ret_list)  
```

```python
m_r_q_table = mean_reward(r_q_table)
m_r_q_network = mean_reward(r_q_network)
m_r_policy = mean_reward(r_policy)

x = np.linspace(1,20000,200)

line1, = plt.plot(x,m_r_q_table, label="Q learning")
line2, = plt.plot(x,m_r_q_network, label="Deep Q Network")
line3, = plt.plot(x,m_r_policy, label="Policy Network")
plt.legend(handles=[line1,line2,line3],loc=4)
plt.ylabel("Average score from last 100 episodes")
plt.xlabel("Number of episodes")
plt.title("Grid Size = 8*8")
```
